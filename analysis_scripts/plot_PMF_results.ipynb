```python
# plot PMF results
# analysis
from gromacs.fileformats.xvg import XVG
import numpy as np
from matplotlib import pyplot as plt
import os



fontsize = 25
plt.rc('font', size=fontsize)
plt.rc('axes', titlesize=fontsize)
plt.rc('axes', labelsize=fontsize) 
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('legend', fontsize=fontsize)

plt.rcParams["font.family"] = "TeX Gyre Heros"


```

```python
# new class
class Result:
    def __init__(self, zArray, zArray2, PMFmean, PMFmean2, PMFstd, PMFstd2):
        self.zArray = zArray
        self.zArray2 = zArray2
        self.PMFmean = PMFmean
        self.PMFmean2 = PMFmean2
        self.PMFstd = PMFstd
        self.PMFstd2 = PMFstd2

    def getResults(self):
        return self.zArray, self.zArray2, self.PMFmean, self.PMFmean2, self.PMFstd, self.PMFstd2

resultList = []
```

```python
path = "/mnt/sdb/xlf/Desktop/ion_channels/TRPV1/TRPV1_1Na/umbrella_sampling_cylin_up/analysis"
os.chdir(path)

# analysis
a = XVG('histo.xvg')
data = a.array
zArray = a.array[0,:] # unit: ps
countData = a.array[1:,:]

a = XVG('profile.xvg')
data = a.array
# zArray = a.array[0,:]
PMF = data[1,:]

isBootstrap=True
if isBootstrap ==True:
    a = XVG('bsResult.xvg')
    data = a.array
    PMFmean = data[1,:]
    PMFstd = data[2,:]

    
    
    
path = "/mnt/sdb/xlf/Desktop/ion_channels/TRPV1/TRPV1_1CAM/umbrella_sampling_cylin_up/analysis"
os.chdir(path)

# analysis
a = XVG('histo.xvg')
data = a.array
zArray2 = a.array[0,:] # unit: ps
countData = a.array[1:,:]

a = XVG('profile.xvg')
data = a.array
# zArray = a.array[0,:]
PMF2 = data[1,:]


isBootstrap=True
if isBootstrap ==True:
    a = XVG('bsResult.xvg')
    data = a.array
    PMFmean2 = data[1,:]
    PMFstd2 = data[2,:]

# compare two results
# plot with error bar
plt.subplots(1,1,figsize = (16,9))
plt.plot(zArray,PMFmean)

plt.ylabel("PMF (kJ/mol)")
plt.title("PMF of ion in TRPV1")
plt.plot(zArray2,PMFmean2)
plt.legend([r'Na$^+$',r'Ca$^{2+}$'])

plt.fill_between(zArray,PMFmean-PMFstd,PMFmean+PMFstd,facecolor = 'blue',alpha = 0.2)
plt.fill_between(zArray2,PMFmean2-PMFstd2,PMFmean2+PMFstd2,facecolor = 'orange',alpha = 0.2)

plt.xlabel(r"COM $\Delta$z (nm)")

ax = plt.gca()
width = 3
ax.spines["top"].set_linewidth(width)
ax.spines["right"].set_linewidth(width)
ax.spines["bottom"].set_linewidth(width)
ax.spines["left"].set_linewidth(width)
plt.tick_params(width=3,length = 10)


result = Result(zArray, zArray2, PMFmean, PMFmean2, PMFstd, PMFstd2)
resultList.append(result)
```

```python
# get results
plt.subplots(3,1,figsize = (10,24))

for i in range(3):
    zArray, zArray2, PMFmean, PMFmean2, PMFstd, PMFstd2 = resultList[i].getResults()
    # plot with error bar
    plt.subplot(3,1,i+1)

    plt.plot(zArray,PMFmean)

    plt.ylabel("PMF (kJ/mol)")
    # plt.title("PMF of ion in TRPV6")
    plt.plot(zArray2,PMFmean2)
    
    if i==2:
        plt.xlim(-3.3,0.7)
        # plt.legend([r'Ca$^{2\!+}$(pulled)+$\rm Na^+$(fixed)',r'Ca$^{2\!+}$(pulled)+Ca$^{2\!+}$(fixed)'])
        plt.legend([r'Ca$^{2\!+}$ with $\rm Na^+$ bound at S1',r'Ca$^{2\!+}$ with Ca$^{2\!+}$ bound at S1'])
    else:
        plt.legend([r'Na$^+$',r'Ca$^{2\!+}$'])
        
    plt.fill_between(zArray,PMFmean-PMFstd,PMFmean+PMFstd,facecolor = 'blue',alpha = 0.2)
    plt.fill_between(zArray2,PMFmean2-PMFstd2,PMFmean2+PMFstd2,facecolor = 'orange',alpha = 0.2)

    plt.xlabel(r"COM $\Delta$z (nm)")
    
        
    
    ax = plt.gca()
    width = 3
    ax.spines["top"].set_linewidth(width)
    ax.spines["right"].set_linewidth(width)
    ax.spines["bottom"].set_linewidth(width)
    ax.spines["left"].set_linewidth(width)
    plt.tick_params(width=3,length = 10)

    # show min PMF results
    print("Minimal energy for Na+ = {:.1f} kJ/mol".format(PMFmean.min()))
    print("Minimal energy for Ca2+ = {:.1f} kJ/mol".format(PMFmean2.min()))

```
